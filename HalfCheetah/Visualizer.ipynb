```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('TrainingData.csv')
plt.hist(df['score'], bins=1000, edgecolor='black')
plt.show()
```

```python
#remove outliers
scores = df['score']
for _ in range(1000):
    mean_score = np.mean(scores)
    std = np.std(scores)
    scores = scores[scores > mean_score - 3.87 * std]
    scores = scores[scores < mean_score + 3.87 * std]
print(len(scores))
```

```python

plt.hist(scores, bins=1000, edgecolor='black')
plt.show()
```
